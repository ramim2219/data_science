## **Mean Deviation using graph**

```python
sec_a = np.array([75,65,73,68,72,76])
sec_b = np.array([90,47,43,96,93,51])
no = np.array([1,2,3,4,5,6])
plt.figure(figsize=(10,3))
plt.scatter(sec_a,no,label="Sec A")
plt.scatter(sec_b,no,color="red",label="Sec B")
plt.plot([70,70,70,70,70,70],no,c="blue",label="Mean")
plt.legend()
plt.show()
```

## **Mean Absolute Deviation**

the mean absolute deviation of a dataset is the average distance between each data point and the mean. It gives us an idea about the variability in a dataset.
Screenshot 2024-07-12 123016.png

```python
sec_a = np.array([75,65,73,68,72,76])
sec_b = np.array([90,47,43,96,93,51])
mn = np.mean(sec_a)
mad_a=np.sum(np.abs(sec_a-mn))/len(sec_a)
mad_b=np.sum(np.abs(sec_b-mn))/len(sec_b)
'''whose mad will less. It will be the ans'''
mad_a,mad_b
```

## **Standard Deviation**


The standard deviation is a measure of the amount of variation or dispersion of a set od values. a low standard deviation indicates that the values tend to be close to the mean (also called the expected value) of the set, while a high standard deviation indicates that the values are spread out over a wider range.

(std)^2 = variance


```python
'''whose std/var will less. It will be the ans'''
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sec_a = np.array([75,65,73,68,72,76])
sec_b = np.array([90,47,43,96,93,51])
```

```python
np.std(sec_a) , np.std(sec_b)
```

```python
np.var(sec_a) , np.var(sec_b)
```
